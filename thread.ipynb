```python
from datetime import datetime
```

```python
df['review_date'] = pd.to_datetime(df['review_date'])

df['year'] = df['review_date'].apply(lambda date:date.year)
df['month'] = df['review_date'].apply(lambda date:date.month)
df['day'] = df['review_date'].apply(lambda date:date.day)
```

```python
df.head()
```

```python
graph = sns.countplot(x='rating', hue='day', data=df).set_title('Daywise rating')
plt.legend([],[], frameon=False)
```

```python
graph = sns.countplot(x='rating', hue='day', data=df).set_title('Daywise rating')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1.5))
```
